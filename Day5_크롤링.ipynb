```python
import requests                    
from bs4 import BeautifulSoup as bs
import pandas as pd

song_data = []
rank = 1

# 전체 노래 100곡 추출

url = "https://music.bugs.co.kr/chart"
html = requests.get(url)
soup = bs(html.text)
songs = soup.select('table.byChart > tbody > tr')

for song in songs :
    title = song.select('p.title > a')[0].text
    singer = song.select('p.artist > a')[0].text
    song_data.append(['Bugs', rank, title, singer])
    rank += 1
    
# 2중 리스트로 만든 100곡을 데이터프레임으로 전환하기

df = pd.DataFrame(song_data, columns = ['서비스', '순위', '노래제목', '가수명'])


# 데이터 프레임을 엑셀 파일로 저장하기

df.to_excel('Bugs.xlsx', index=False)
```

```python
# 멜론

from selenium import webdriver          # 브라우저 연결
from bs4 import BeautifulSoup as bs
import pandas as pd

song_data = []
rank = 1

driver = webdriver.Chrome('c:/Users/EZEN/chormedriver.exe')  # 크롬드라이버 위치 가져오기
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)
html = driver.page_source
soup = bs(html)
songs = soup.select('tbody > tr')

for song in songs :
    title = song.select('div.rank01 > span > a')[0].text
    singer = song.select('div.rank02 > span > a')[0].text
    song_data.append(['Melon', rank, title, singer])
    rank += 1

df = pd.DataFrame(song_data, columns = ['서비스', '순위', '곡명', '가수명'])
df
```
